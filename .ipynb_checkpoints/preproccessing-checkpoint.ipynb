```python
import re
from nltk.stem import PorterStemmer

#stemmer creation
stemmer = PorterStemmer()

def load_stopwords(path=r'C:\Users\Ibad Ur Rehman\Desktop\Sem 6\IR\VSM_IR\corpus\Stopword-List (1).txt'):
    #loading stopwords
    stopwords = set()
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            word = line.strip().lower()
            if word:  #empty lines skipping arc
                stopwords.add(word)
    return stopwords

# globally load kar lia taake hr call pe file na kholni paray
STOPWORDS = load_stopwords()

def preprocess(text):
    # lowercase -> TRUMP aur trump same hone chahiye
    text = text.lower()

    # keep alphabets only, numbers aur punctuation khatm krdo. that's basically a regular expression, what we learnt in TOA
    tokens = re.findall(r"[a-z']+", text)

    # stopwords hata diey
    tokens = [t for t in tokens if t not in STOPWORDS]

    #now apply stemming on tokens
    tokens = [stemmer.stem(t) for t in tokens]

    return tokens
```
